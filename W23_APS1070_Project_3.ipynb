````python
### YOUR CODE HERE ###

# comput the covariance matrix of data
n, d = data.shape    
df_cov = np.dot(data.T,data) / (n-1) # covariance matrix, dimension is (483, 483)
````

````python
#define a function returns sorted eigenvalues and eigenvectors
def get_sorted_eigen(df_cov):
  eigenValues, eigenVectors = np.linalg.eigh(df_cov)

  # sort eigenVectors in descending order by their eigenValue
  idx = np.argsort(-eigenValues)
  eigenValues = eigenValues[idx]
  eigenVectors = eigenVectors[:, idx]

  return eigenValues, eigenVectors
````

````python
# effectiveness of your principal components in covering the variance of the dataset
eigenValues, eigenVectors = get_sorted_eigen(df_cov)
eigValSum = sum(eigenValues)
expVar = [eigV/eigValSum*100 for eigV in eigenValues]
cumExpVar = np.cumsum(expVar)

# Plot the scree plot with variance
plt.plot(cumExpVar, 'r-o', label='Cumulative Explained Variance')
plt.plot(expVar, 'b-o', label='Cumulative Explained Variance')
plt.legend()
plt.show()
````

````python
# number of PCs needed to cover 99% of the dataset's variance
pc_num = np.argmax(cumExpVar >= 99) + 1
print("number of PCs needed to cover 99% of the dataset's variance is", pc_num)
````

````python
from sklearn.metrics import mean_squared_error
````

````python
def rmse(yPred, y):
    return np.sqrt(mean_squared_error(yPred, y))
````

````python
def plot_country_figures(original_df, country_name):
  ### YOUR CODE HERE ### 

  #plot1: original time-series for the specified country.
  plt.subplot(4, 1, 1)
  original_df.loc[country_name,:].plot(figsize = (15,15))
  plt.title("original time-series for "+ country_name, y=1.0, pad=-14)


  #plot2: 5 curves incremental reconstruction of the original time-series
  # PC1 only, PC1-2, PC1-4, PC1-8, PC1-16
  plt.subplot(4, 1, 2)

  #standardized original_df
  scaler = StandardScaler()
  std_df = scaler.fit_transform(original_df)
  std_df = pd.DataFrame(std_df, columns = original_df.columns, index = original_df.index)

  #eigenvectors and eigenvalues of std_df
  n_sd, d_sd = std_df.shape    
  df_cov_sd = np.dot(std_df.T,std_df) / (n_sd-1) 
  eigenValues_sd, eigenVectors_sd = get_sorted_eigen(df_cov_sd)

  res_all = []
  pc_list = [1,2,4,8,16]
  #projections
  for pc in pc_list:
    w = eigenVectors_sd[:, 0:pc]
    projX = np.dot(std_df.loc[country_name,:], w)
    reconX = np.dot(projX.reshape(1,-1), w.T)

    #transfer back
    reconX = scaler.inverse_transform(reconX)
    reconX_df = pd.DataFrame(reconX, columns = original_df.columns)

    #residual error of the current curve
    res = original_df.loc[country_name,:].to_numpy() - reconX
    res_all.append(res)

    #plot the curve
    reconX_df.iloc[0,:].plot(figsize = (15,15), label = "PC1 to PC"+str(pc))
  plt.legend()
  plt.title("incremental reconstruction of the original time-series (5 curves) - " + country_name)
  plt.show()


  # plot3: The residual error for your best reconstruction 
  plt.subplot(4, 1, 3)

  # check the best reconstruction among all 5 curves
  sum_res = [np.mean(np.sum(element ** 2)) for element in res_all]
  min_res_index = np.argmin(sum_res) # index for best construction
  min_res = res_all[min_res_index]
  
  #plot the residual error
  minres_df = pd.DataFrame(min_res, columns = original_df.columns)
  minres_df.iloc[0,:].plot(figsize = (15,15), label = "PC1 to PC"+str(pc_list[min_res_index]))
  plt.legend()
  plt.title("Residual Error for best construction - "+country_name)

  
  #plot 4: The RMSE of the reconstruction as a function of the number of included components.(x:1-10)
  plt.subplot(4, 1, 4)

  rmse_all = []
  for x in range(1, 11):
    w1 = eigenVectors_sd[:, 0:x]
    projX1 = np.dot(std_df.loc[country_name,:], w1)
    reconX1 = np.dot(projX1.reshape(1,-1), w1.T)

    #transfer back
    reconX1 = scaler.inverse_transform(reconX1)
    reconX_df1 = pd.DataFrame(reconX1, columns = original_df.columns)

    #rmse
    rmse1 = rmse(reconX_df1.iloc[0,:].to_numpy(), original_df.loc[country_name,:].to_numpy())
    rmse_all.append(rmse1)

  rmse_df = pd.DataFrame([rmse_all], columns=list(range(1,11)))
  rmse_df.iloc[0,:].plot(figsize = (15,15))
  plt.xlabel("Number of principal components")
  plt.title("The RMSE of the reconstruction (x from 1 to 10) - "+country_name)
````

Standardization is not required for SVD. Unlike PCA, SVD does not make any assumptions about the data distribution. Besides, PCA needs covariance matrix on computing variances and standardization is necessary for covariance computation, while covariance matrix is not necessary for SVD. Hence SVD does not require standardization.



````python
# comparison on PCA and SVD results (rmse score)
def comparision(original_df, country_name):

  #eigenvectors and reconstruction using PCA
  #standardized original_df
  scaler = StandardScaler()
  std_df = scaler.fit_transform(original_df)
  std_df = pd.DataFrame(std_df, columns = original_df.columns, index = original_df.index)

  #eigenvectors and eigenvalues of std_df(PCA)
  n_sd, d_sd = std_df.shape    
  df_cov_sd = np.dot(std_df.T,std_df) / (n_sd-1) 
  eigenValues_sd, eigenVectors_sd = get_sorted_eigen(df_cov_sd)

  # eigenvectors and eigenvalues of original_df(SVD)
  U, s, V = np.linalg.svd(original_df, full_matrices=False) #s - eigenValues
  eigenVectors = V.T
  idx = np.argsort(-s)
  s = s[idx]
  eigenVectors = eigenVectors[:,idx]

  # PCA rmse score
  rmse_allp = []
  # SVD rmse score
  rmse_alls = []

  #projections
  for pc in range(original_df.shape[1]):

  # PCA projections
    wp = eigenVectors_sd[:, 0:pc]
    projXp = np.dot(std_df.loc[country_name,:], wp)
    reconXp = np.dot(projXp.reshape(1,-1), wp.T)

    #transfer back
    reconXp = scaler.inverse_transform(reconXp)
    reconXp_df = pd.DataFrame(reconXp, columns = original_df.columns)

    # rmse score
    rmse1p = rmse(reconXp_df.iloc[0,:].to_numpy(), original_df.loc[country_name,:].to_numpy())
    rmse_allp.append(rmse1p)

  # SVD projections
    ws = eigenVectors[:, 0:pc]
    projXs = np.dot(original_df.loc[country_name,:], ws)
    reconXs = np.dot(projXs.reshape(1,-1), ws.T) 

    reconXs_df = pd.DataFrame(reconXs, columns = original_df.columns)

    #rmse score
    rmse1s = rmse(reconXs_df.iloc[0,:].to_numpy(), original_df.loc[country_name,:].to_numpy())
    rmse_alls.append(rmse1s)
  
  # return the minimum rmse score for both PCA and SVd
  print("The rmse score for PCA is: ", min(rmse_allp), "for best reconstruction: PC1 - PC"+str(rmse_allp.index(min(rmse_allp))))
  print("The rmse score for SVD is: ", min(rmse_alls), "for best reconstruction: PC1 - PC"+str(rmse_alls.index(min(rmse_alls))))
````

````python
comparision(cases_raw, "United States")
````

````python
comparision(cases_raw, "Canada")
````

By comparing the rmse score for PCA and SVD on best reconstuctions of US, the rmse score of SVD is smaller than PCA. This may because standardization on PCA can amplify the importance of outliers, causing the reconstruction error larger due to different eigenvectors than SVD. 

Besides, the number of principal component in PCA is greater than SVD. This may because PCA and SVD have different goals and PCA aims to maximize the variance of data explained.
