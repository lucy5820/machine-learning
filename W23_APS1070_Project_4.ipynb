````python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from scipy import linalg
````

````python
# split the dataset

target_data = df['critical_temp']
feature_data = df.drop(columns=['critical_temp'])

# 80% train, 20% val
X_train,X_val, y_train, y_val= train_test_split(feature_data, target_data, test_size=0.2, random_state=1)
````

````python
# standardize the data manually
# z = (x - u) / s, where u is the mean and s is the standard deviation of training samples 
def manual_std(df):

    # comput the mean and standard deviation of samples
    mean = df.mean()
    sd = df.std()
    
    # Standardize
    df_std = (df - mean) / sd
    
    return df_std

# manually standardized the training and validation dataset
X_train = manual_std(X_train)
X_val = manual_std(X_val)
````

````python
#Insert a first column of all  1 s in both the training and validation data sets.
X_train.insert(loc = 0, column = "bias", value = 1)
X_val.insert(loc = 0, column = "bias", value = 1)
````

Adding the column of 1s will provide us the bias to the linear regression model. If we adding 1s prior the standardization, then the stardarization process would subtract the 1s column, leading the model shifting but not changing scaling. Hence the 1s column would no longer represent the bias term. Therefore, we add column of 1s after standardization in order to include the bias term.

## Part 2 - Linear Regression Using Direct Solution [1 marks]

Implement the direct solution of the linear regression problem on the training set. **[0.5]**

**Note: You should use scipy.linalg.inv to perform the inversion, as numpy.linalg.inv may caue an error.**

Report the root-mean-square error (RMSE) for both the training and validation sets. **[0.5]**

You may use `mean_squared_error` from Sklearn for computing the RMSE.

````python
### YOUR CODE HERE ###

# direct solution
W = np.dot(np.linalg.inv(np.dot(X_train.T, X_train)), np.dot(X_train.T, y_train))
W[0],W[1]
````

````python
W
````

````python
# Report the root-mean-square error 
def rmse(yPred, y):
    return np.sqrt(mean_squared_error(yPred, y))


y_train_pred = np.dot(X_train, W)
y_val_pred = np.dot(X_val, W)

print('RMSE for training set:', rmse(y_train_pred, y_train))
print('RMSE for validation set:', rmse(y_val_pred, y_val))
````

## Part 3 - Full Batch Gradient Descent [3 marks]

We will now implement a "full batch" gradient descent algorithm and record the training time for our model. Recall that the full batch gradient descent is
 $$w_t = w_{t-1} - \alpha~g_t$$ where $\alpha$ is the learning rate and $g_t$ is your gradient, computed on the entire data. 

Here are the steps for this part:

* Implement gradient descent for linear regression using a fixed learning rate of $\alpha= 0.01$, and iterate until your model's **validation** RMSE converges. 

  We consider the gradient descent as having converged when RMSE on the validation set using gradient descent satisfies:

  $$ RMSE_\text{GD} \leq 1.001 \times RMSE_\text{Direct Solution}$$

  where $RMSE_\text{Direct Solution}$ is the RMSE on the validation set using the direct solution that you calculated in the previous part. 

  We refer to the quantity $RMSE_\text{Direct Solution}\times 1.001$ as the convergence threshold (CT).

*  Record the training time (from the first iteration until convergence) using the `time.time()` function. Be sure to compute the gradients yourself! Take a look at the code provided in the tutorial. **[1]**

* Plot the training RMSE and the validation RMSE vs. epoch on the same figure.  **[1]**

* Comment on overfitting/underfitting by observing the training and validation RMSE **[1]**

**Hint**: Initialize your weights with small random numbers (<$0.001$)

````python
import time
start_time = time.time() ## Records current time

## GD Script -- Sample code in tutorial! ##

w = (0.001 * np.random.random(X_train.shape[1])).reshape(1,-1) ### Inital weights for training dataset

lr = .01  ### Learning rate
rmse_train = []
rmse_val = []

threshold = 1.001 * rmse(y_val_pred, y_val)

update = 100 
count = 0

while update > threshold:
  # training set
  y_train_gd = np.dot(X_train,w.T).reshape(-1,1)
  rmse_train.append(rmse(y_train_gd,y_train))
  gradient_train = (1/len(y_train_gd) * np.dot(X_train.T, y_train_gd - y_train.values.reshape(-1, 1))).reshape(1,-1)
  

  w = w - lr * gradient_train ### weight update

  # validation set
  y_val_gd = np.dot(X_val,w.T).reshape (-1,1)
  rmse_val.append(rmse(y_val_gd,y_val))
  
  # check for threshold
  update = rmse(y_val_gd, y_val)
  count += 1

print("the total number of epoch:", count)

print("--- Total Training Time: %s (s) ---" % (time.time() - start_time))
````

````python
plt.figure(figsize=(10,5))
plt.plot(rmse_train, label = "training set")
plt.plot(rmse_val, label = "validation set")

plt.xlabel("Epoch")
plt.ylabel("RMSE")

plt.legend()
plt.show()
````

The model is neither overfitting nor underfitting, since the difference of  RMSE between training set and validation set is quite small for each epoch and the final RMSE is relatively low. 

If the model is overfitting, then the RMSE is likely to be low for training set but high for validation set. If the model is underfitting, RMSE for both the sets are likely to be high.



## Part 4 - Mini-batch and Stochastic Gradient Descent [4 marks]

Write a function that performs mini-batch gradient descent until the convergence threshold (CT) is reached. **[1]**

The inputs of that function are: 
  * Input data (training/validation), Batch size, Learning Rate, Convergence Threshold (CT)

Your function will return the following arrays: 
 * The final weights after training.
 * The training RMSE at each epoch.
 * The validation RMSE at each epoch.
 * An array that has the elapsed time from the start of the training process to the end of each epoch (e.g., if each epoch takes exactly 2 seconds, the array would look like: [2 4 6 8 ...]).

Notes: 
* For certain batch sizes, GD might not converge to a solution. For that reason, you need to check the RMSE of the validation/training set at each epoch, and if it's getting larger and larger, you should stop the training for that case (the design is up to you!).
* CT will help you to know when your model is converged. 
* **Important: after each epoch, you need to shuffle the entire training set.** This ensures that new mini-batches are selected for every epoch. Hint: use `np.random.permutation`. 

Let's now use the function to investigate the effect of batch size on convergence. When the batch size is 1, we call that stochastic gradient descent. When the batch size equals the # of training data, it is full-batch (i.e., all data points are used at every iteration). Anywhere in between is mini-batch (we use some of the data).
  * Sweep different values for the mini-batch size (at least 5 values that **converge**), each time using a learning rate of $\alpha= 0.01$. **Hint: Try batch sizes that are powers of two (e.g., 2,4,8,16,32,64,128...).** These batch sizes fit better on the hardware and may achieve higher performance.  **[0.5]**

Provide the following $3$ plots:
  1.  Plot training and validation RMSE vs. **epoch** for all the **converging** batch sizes (some batch sizes might not converge) in a figure. The X-axis is Epoch # and the Y-axis is RMSE. **[0.5]**
  2. Plot training and validation RMSE vs. **time** for all the **converging** batch sizes in a figure. The X-axis is Time, and the Y-axis is RMSE. **[0.5]**
  3. Plot Total training time (y-axis) vs. Batch size (x-axis). **[0.5]**

Describe your findings, including the main takeaways from each of your plots. **[1]**

````python
### YOUR CODE HERE ###
import time

def mini_gd(X_train, y_train, X_val, y_val, bs, lr, ct):
  # return final weights, each epoch training RMSE, each epoch validation RMSE
  rmse_train = []
  rmse_val = []
  epoch_time = []

  w = (0.001 * np.random.random(X_train.shape[1])).reshape(1,-1) ## initial weights for training dataset # w_train.shape = (1, 82)

  update = ct*bs+1 # make sure the loop starts, will update during the while loop
  count = 0
  converge = True

  start_time = time.time()


  while converge and update > ct:
    # training set
    # select mini-batch of training data
    train_all = X_train.copy()
    train_all["critical_temp"] = y_train

    train_idx = np.random.permutation(train_all.shape[0])[:bs] # shuffle each mini-batch
    mini_train = train_all.iloc[train_idx,:]

    y_train_mini = mini_train['critical_temp'] 
    X_train_mini = mini_train.drop(columns=['critical_temp'])  #shape: bs, 82
    
    # compute and update the gradient
    y_train_gd = np.dot(X_train_mini,w.T).reshape(-1,1)
    gradient_train = (1/len(y_train_gd) * np.dot(X_train_mini.T, y_train_gd - y_train_mini.values.reshape(-1, 1))).reshape(1,-1)
    w = w - lr * gradient_train ### weight update


    # training RMSE
    y_train_pred = np.dot(X_train,w.T).reshape(-1,1)
    rmse_train.append(rmse(y_train_pred,y_train))

    # val RMSE
    y_val_pred = np.dot(X_val,w.T).reshape(-1,1)
    rmse_val.append(rmse(y_val_pred,y_val))

    # check for threshold
    update = rmse_val[-1]
    epoch_time.append(time.time() - start_time)

    # check for divergence
    if count > 2:
        if rmse_val[-1] > rmse_val[-2] > rmse_val[-3] > rmse_val[-4]: # if the rmse keep increasing for 3 epochs
              print("The model is diverging at batch size", bs)
              converge = False

    count += 1
    
  if converge:
    print("The model is converging at batch size", bs)

    
  print("total epochs:", count)
  # return final weights, each epoch training RMSE, each epoch validation RMSE
  return w, rmse_train, rmse_val, epoch_time, epoch_time[-1]
````

````python
lr = .01  ### Learning rate
ct = 1.001 * rmse(y_val_pred, y_val)

test_bs = [2,4,8,16,32,64,128, 256, 512, 1024]
for i in test_bs:
  w, rmse_train, rmse_val, epoch_time, total_time = mini_gd(X_train, y_train, X_val, y_val, i, lr, ct*1.2)
````

The model is converaging at batch size: 64, 128, 256, 512, 1024

````python
#Plot training and validation RMSE vs. epoch for all the converging batch sizes in a figure. The X-axis is Epoch # and the Y-axis is RMSE. [0.5]

# training and validation RMSE vs epoch
batch_size = [64, 128, 256, 512, 1024]

results = {}

for bs in batch_size:
  w, rmse_train, rmse_val, epoch_time, total_time = mini_gd(X_train, y_train, X_val, y_val, bs, lr, ct* 1.2)
  results[bs] = {'w': w, 'rmse_train': rmse_train, 'rmse_val': rmse_val, 'epoch_time': epoch_time, 'total_time': total_time}
  
````

````python
for i in batch_size:
  plt.title('training and validation RMSE vs epoch')
  plt.plot(results[i]["rmse_train"], label = 'Training Batch Size {}'.format(i), linestyle = 'dashed') 
  plt.plot(results[i]["rmse_val"], label = 'Validation Batch Size {}'.format(i)) 
  plt.xlabel('Epoch')
  plt.ylabel('RMSE')
  plt.legend()
````

````python
#Plot training and validation RMSE vs. time for all the converging batch sizes in a figure. The X-axis is Time, and the Y-axis is RMSE. [0.5]

for i in batch_size:
  plt.title('training and validation RMSE vs time')
  plt.plot(results[i]["epoch_time"], results[i]["rmse_train"], label = 'Training Batch Size  {}'.format(i), linestyle = 'dashed') 
  plt.plot(results[i]["epoch_time"], results[i]["rmse_val"], label = 'Validation Batch Size {}'.format(i)) 
  plt.xlabel('Time')
  plt.ylabel('RMSE')
  plt.legend()
````

````python
#Plot Total training time (y-axis) vs. Batch size (x-axis). 
test_bs = [2,4,8,16,32,64,128, 256, 512, 1024]

total_time_all = []
for i in test_bs:
  w, rmse_train, rmse_val, epoch_time, total_time = mini_gd(X_train, y_train, X_val, y_val, i, lr, ct*1.2)
  total_time_all.append(total_time)
plt.title('Total training time vs Batch size')
plt.plot(test_bs, total_time_all) 
plt.xlabel('Batch Size')
plt.ylabel('Total training time')
````
